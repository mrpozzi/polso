```python
date_columns = [ 'Join Date', 'Birth Date', 'Termination Date']

def convert_to_date(raw_date):
    date = datetime.strptime(raw_date, '%m/%d/%Y')

def _parse_wg_date(ds):
    
    for fmt in ('%m/%d/%y', '%m/%d/%Y', '%m/%d//%Y', '%m/%d/ %Y'):
        try:
            return(datetime.strptime(ds, fmt))
        except ValueError:
            pass
        except TypeError:
            return('No Date')
    try:
        if len(ds.split(' ')) > 1:
            return(datetime.strptime(ds.split(' ')[1], '%m/%d/%Y'))
    except AttributeError:
        print(ds)
        return('No Date')
    except ValueError:
        print(ds)
        return('No Date')
    
    raise ValueError('no valid date format found for ' + ds)
```
